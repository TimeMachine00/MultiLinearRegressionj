```python
import pandas as pd
```

```python
loadDataset = pd.read_csv("C:\\Users\\HussainSha\\ML_forGithub\\Multi_linear_regression\\insurance.csv")
loadDataset.head(10)
```

```python
dummies = pd.get_dummies(loadDataset[['sex', 'smoker', 'region']])
```

```python
dummies
```

```python
new_dummies = dummies.drop('region_southwest', axis='columns')
```

```python
new_dataset = loadDataset.drop(['sex', 'smoker', 'region'], axis='columns')
```

```python
finalDataset = pd.concat([new_dataset,new_dummies],axis='columns' )
```

```python
finalDataset.head(15)
```

```python
x= finalDataset.drop(['charges'], axis='columns')
x.head(5)
```

```python
y=finalDataset['charges']
```

```python
y.head(5)
```

```python
from sklearn.model_selection import train_test_split
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(x,y, test_size=0.2, random_state=10)
```

```python
xTrain
```

```python
from sklearn import linear_model
```

```python
Multi_lin_reg = linear_model.LinearRegression()
```

```python
Multi_lin_reg.fit(xTrain, yTrain)
```

```python
Multi_lin_reg.coef_
```

```python
Multi_lin_reg.intercept_
```

```python
Multi_lin_reg.predict(xTest)
```

```python
Multi_lin_reg.score(xTest,yTest)
```
